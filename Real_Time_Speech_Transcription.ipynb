```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition", 
             model="abdouaziiz/wav2vec2-xls-r-300m-wolof-lm")
#p = pipeline("automatic-speech-recognition")
```

```python
import gradio as gr
```

```python
import time

def transcribe(audio, state=""):
    time.sleep(3)
    text = p(audio)["text"]
    state += text + " "
    return state, state
```

```python
gr.Interface(
    fn=transcribe,
    inputs = [
        gr.inputs.Audio(source="microphone", type="filepath"),
        "state"
    ],
    outputs = [
        "textbox",
        "state"
    ],
    live=True
).launch()
```
